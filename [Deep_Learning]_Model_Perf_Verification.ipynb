## 학습셋과 테스트셋

과적합 방지의 한 방법

```python
# 학습셋, 테스트셋 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
# 모델 재학습
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
# 앞서 학습한 모델 테스트셋에 적용
score = model.evaluate(X_test, y_test) # evaluate() : 모델 평가 함수(loss와 accuracy 반환)
print('Test accuracy :', score[1])
```

딥러닝, 머신 러닝의 목표는 새로운 데이터에 대해 높은 정확도를 안정되게 보여주는 모델을 만드는 것<br>

모델 성능 향상을 위한 방법에는 크게 두 가지(데이터를 보강하는 방법, 알고리즘을 최적화하는 방법)<br>

데이터를 이용해 성능을 향상시키려면,<br>
1. 충분한 양의 데이터 준비
2. (데이터 양이 충분하지 않을 경우) 데이터 증강(사진 크기 확대/축소, 사진 위치 변경(위아래로))
3. 테이블형 데이터의 경우 이상치가 모델에 영향을 못 끼치도록 값의 크기를 적절히 조절(표준화, 정규화)
4. 교차 검증

알고리즘을 이용해 성능을 향상시키려면,
- 다른 구조로 모델을 바꿔가며 최적의 구조 찾기(은닉층 개수, 노드의 수, 최적화 함수 종류 등)

```python
# 전체 코드
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

import pandas as pd

!git clone https://github.com/taehojo/data.git

df = pd.read_csv('./data/sonar3.csv', header=None)

X = df.iloc[:, :60]
y = df.iloc[:, 60]

k = 5

kfold = KFold(n_splits=k, shuffle=True)

acc_score = []

def model_fn():
  model = Sequential()
  model.add(Dense(24, input_dim=60, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  return model

for train_index, test_index in kfold.split(X):
  X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  model = model_fn()
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  history = model.fit(X_train, y_train, epochs=200, batch_size=10)

  accuracy = model.evaluate(X_test, y_test)[1]
  acc_score.append(accuracy)

avg_acc_score = sum(acc_score) / k

print('Accuracy :', acc_score)
print('Accuracy average :', avg_acc_score)
```
